```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.naive_bayes import GaussianNB 
```

```python
df =pd.read_csv("healthcare-dataset.csv") 
df
```

```python
df['gender']=df['gender'].map({'Male': 0,'Female': 1}) 
df['ever_married']=df['ever_married'].map({'Yes': 0,'No': 1}) 

df['work_type']=df['work_type'].map({'Private':0,'Self-employed':1,'Govt_job':2,'children':3,'Never_worked':4}) 

df['Residence_type']=df['Residence_type'].map({'Rural':0,'Urban':1}) 

df['smoking_status']=df['smoking_status'].map({'never smoked':0,'formerly smoked':1}) 
df.dropna(inplace=True) 
```

```python
X=df.iloc[:,1:11] 
Y=df.iloc[:,11] 
```

```python
X
```

```python
Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.20, random_state = 45) 
```

```python
classifier = GaussianNB() 
classifier.fit(X_train, y_train) 
GaussianNB()
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
```
